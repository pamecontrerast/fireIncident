##### Crear tabla de hechos y dimensiones (DWH)

```python
df = pd.read_parquet("/Users/pamelacontreras/Desktop/tenpo/files/Fire_Incidents-2023-12-03.parquet")
```

```python
# se crea archivo vacio, simil a crear tabla inicial de
pd.DataFrame([], columns=['id_battalion', 'battalion', 'station_area', 'city', 'box']).to_parquet("files/dim_battalion.parquet")
```

```python
def assign_battalion(df):
    df_tbl_dim_battalion = pd.read_parquet("files/dim_battalion.parquet")
    #se intenta asignar batallon
    df = df.merge(df_tbl_dim_battalion, how="left", on=['battalion','station_area','city','box'])
    df_sin_batallon = df[df.id_battalion.isna()]#[["battalion", "station_area", "city", "box"]]
    if(len(df_sin_batallon)>0):
        df = df.drop(columns=["id_battalion"])
        df_tbl_dim_battalion_sin_id = df_sin_batallon[["battalion", "station_area", "city", "box"]].drop_duplicates()
        df_tbl_dim_battalion_sin_id = df_tbl_dim_battalion_sin_id.reset_index(drop=True).reset_index().rename(columns={'index':'id_battalion'})
        #se mueve segun el largo del batallon guardado
        df_tbl_dim_battalion_sin_id.id_battalion = df_tbl_dim_battalion_sin_id.id_battalion + len(df_tbl_dim_battalion)
        df_tbl_dim_battalion = pd.concat([df_tbl_dim_battalion, df_tbl_dim_battalion_sin_id])
        df = df.merge(df_tbl_dim_battalion, how="left", on=['battalion','station_area','city','box'])
        df_tbl_dim_battalion.to_parquet("files/dim_battalion.parquet")
        df = df.drop(columns=['battalion','station_area','city','box'])
    return df 
```

```python
print(pd.read_parquet("files/dim_battalion.parquet").shape)
df1 = assign_battalion(df)
print(df1.shape)
print(pd.read_parquet("files/dim_battalion.parquet").shape)
df2 = assign_battalion(df)
print(df2.shape)
print(pd.read_parquet("files/dim_battalion.parquet").shape)
```
